```python
import numpy as np
import pandas as pd
```

```python
target = np.random.normal(10, 10, 1000)
df = pd.DataFrame(target, columns=['target'])

df.head()
```

```python
def format_X_Y(dataset, x_size, y_size):
    X, Y = [], []
    N = len(dataset) - x_size - y_size + 1
    for idx in range(N):
        _X = dataset[idx:idx + x_size]
        X.append(_X)

        _Y = dataset[idx + x_size:idx + x_size + y_size]
        Y.append(_Y)
    X, Y = np.array(X).squeeze(-1), np.array(Y).squeeze(-1)
    return X, Y
```

```python
dataset = df.to_numpy()
```

```python
x_size, y_size = 10, 1

X, Y = format_X_Y(dataset, x_size, y_size)
df = pd.DataFrame(
    np.concatenate([X, Y], axis=1),
    columns=(
        [f'feature_{i}' for i in range(1, x_size + 1)]
        + [f'target_{i}' for i in range(1, y_size + 1)]
    )
)
df
```

```python
x_size, y_size = 10, 3

X, Y = format_X_Y(dataset, x_size, y_size)
df = pd.DataFrame(
    np.concat([X, Y], axis=1),
    columns=(
        [f'feature_{i}' for i in range(1, x_size + 1)]
        + [f'target_{i}' for i in range(1, y_size + 1)]
    )
)
df
```

```python
feature_1 = np.random.normal(1, 1, 1000)
feature_2 = np.random.normal(10, 1, 1000)
feature_3 = np.random.normal(100, 1, 1000)
target = np.random.normal(1000, 1, 1000)


df = pd.DataFrame(np.stack([feature_1, feature_2, feature_3, target], axis=1),
                  columns=['feature_1', 'feature_2', 'feature_3', 'target'])
```

```python
target_col = 'target'
feature_cols = ['feature_1', 'feature_2', 'feature_3']

target_dataset = df.loc[:, [target_col]].to_numpy()
feature_dataset = df.loc[:, feature_cols].to_numpy()
```

```python
def format_target_X_Y(target_dataset, target_x_size, target_y_size):
    X, Y = [], []
    N = len(target_dataset) - target_x_size - target_y_size + 1
    for idx in range(N):
        _X = target_dataset[idx:idx + target_x_size]
        X.append(_X)

        _Y = target_dataset[idx + target_x_size:idx + target_x_size + target_y_size]
        Y.append(_Y)
    X, Y = np.array(X).squeeze(-1), np.array(Y).squeeze(-1)
    return X, Y


def format_feature_X_Y(feature_dataset, target_x_size, target_y_size, feature_x_size):
    X = []
    N = len(feature_dataset) - target_x_size - target_y_size + 1
    for idx in range(N):
        _X = feature_dataset[idx:idx + target_x_size][-feature_x_size:].T
        X.append(_X)
    X = np.array(X)
    X = X.reshape(X.shape[0], -1)
    return X
```

```python
target_x_size, target_y_size = 3, 1
feature_x_size = 1

target_X, target_Y = format_target_X_Y(target_dataset, target_x_size, target_y_size)
feature_X = format_feature_X_Y(feature_dataset, target_x_size, target_y_size, feature_x_size)
```

```python
columns = []
for feature_col in feature_cols:
    columns.extend(
        [f'{feature_col}_{idx}' for idx in range(1, feature_x_size + 1)]
    )
columns += [f'feature_{i}' for i in range(1, target_x_size + 1)]
columns += [f'target_{i}' for i in range(1, target_y_size + 1)]

pd.DataFrame(
    np.concatenate([feature_X, target_X, target_Y], axis=1),
    columns=(
        columns
    )
)
```
